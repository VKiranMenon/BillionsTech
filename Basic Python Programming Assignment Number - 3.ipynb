Question 1. 

Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
insert_num = float(input("Please enter a number: "))

if insert_num > 0:
    print("The number inserted ", insert_num, " is positive. ")
    
elif insert_num < 0:
    print("The number inserted ", insert_num, " is negative. ")
    
elif insert_num == 0:
    print("The number inserted ", insert_num, " is 0. ")

```

Question 3. 

Write a Python Program to Check Leap Year?

```python
## Method 1

insert_year = int(input("Please enter a year. "))

if type(insert_year) == float:
    print("Please enter a year without decimal. ")
    
elif insert_year < 0:
    print("Year cannot be negative. ")
    

    
elif insert_year % 4 == 0 and insert_year % 100 != 0 or insert_year % 400 == 0 :
    print("The year entered is a leap year. ")
    
elif insert_year % 4 != 0 and insert_year % 100 == 0 or insert_year % 400 != 0:
    print ("The year entered is NOT a leap year. ")
```

```python
## Method 2

## Coding Exercise Leap Year

check_year = int(input("Enter the year: "))

if check_year % 4 == 0:
    
    if check_year % 100 == 0:
        
        if check_year % 400 == 0:
            print (check_year, "is a Leap Year. ")
            
        else:
            print (check_year, "is NOT a Leap year")
            
    else:
        print (check_year, "is NOT a Leap year")
        
else:
    print (check_year, "is NOT a Leap year")
```

Question 4. 

Write a Python Program to Check Prime Number?

```python
## Method 1

num = int(input("Enter a number : "))

for i in range (2, num):
    if num % i == 0:
        
        print("Not Prime ")
        break
        
else:
    print("Prime")
```

```python
## Method 2

def find_prime():
    num = int(input("Enter a number : "))
    
    for i in range (2, num):
        if num % i == 0:
            print("Not Prime ")
            break
            
    else:
        print("Prime ")            
    return  
```

And that any of that number is the divisor will give NO remainder becomes a common Number...

And any number which is divided by the divisior which is the  variable 'i' gives a remainder becomes a PRIME number.
